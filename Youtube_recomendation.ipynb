```python
def clean_text(text):
  text = re.sub(r'@[A-Za-z0-9]+','',text) #removes @mentions
  text = re.sub(r'#','',text) #removes # tags
  text = re.sub(r'\\n','',text) #removes # tags
  text = re.sub(r'RT[\s]+','',text) #removes RT
  text = re.sub(r'https?:\/\/\S+','',text) #removes urls

  return text
youtube_data['clean_text']=youtube_data['Title'].apply(clean_text)

#display clean text
youtube_data.head(5)
```

```python
tf_idf = TfidfVectorizer(stop_words='english')

tf_idf_matrix = tf_idf.fit_transform(youtube_data['Title']);
print(tf_idf_matrix)


similarity = cosine_similarity(tf_idf_matrix)
```

```python
def news_recommendation(Title, similarity, indices, youtube_data):
    if title not in indices:
        print(Title)
        return "Title not found in the dataset."

    index = indices[title]
    print(index)
    similarity_scores = list(enumerate(similarity[index]))
    print(similarity_scores)
    similarity_scores = sorted(similarity_scores, key=lambda x: x[1], reverse=True)

    similarity_scores = similarity_scores[:10]
    print(similarity_scores)
    news_indices = [i[0] for i in similarity_scores]
    print(news_indices)

    return youtube_data['Title'].iloc[news_indices]
```
